```python
# Obi - Ignore all of this computer code just read the first few lines in green.
# <-- play button here
# All you need to do is press the play button
# after you press play, it will ask you at the bottom to upload your webpage url
# just copy and paste the webpage that you are wanting the QR code for
# it will make the QR code and give you the option to make another

import sys
!pip install qrcode

import qrcode
import os
from IPython.display import Image, display

def generate_qr_code(url):
    """Generates a QR code for the given URL and saves it to a file."""
    try:
        # Create QR code instance
        qr = qrcode.QRCode(
            version=1,  # Controls the size of the QR Code
            error_correction=qrcode.constants.ERROR_CORRECT_L, # About 7% or less errors can be corrected
            box_size=10, # How many pixels each "box" of the QR code will be
            border=4, # How many boxes thick the border around the QR code will be
        )
        qr.add_data(url)
        qr.make(fit=True)

        # Create an image from the QR Code instance
        img = qr.make_image(fill_color="black", back_color="white")

        # Generate a filename from the URL, replacing non-alphanumeric chars with underscores
        # and ensuring it ends with .png
        safe_url_name = "".join(c for c in url if c.isalnum() or c in ['.', '_', '-']).replace('/', '_').replace(':', '_').replace('?', '_').replace('=', '_')
        filename = f"qr_code_{safe_url_name}.png"
        # Ensure the filename is not excessively long
        if len(filename) > 200:
            filename = filename[:200] + '.png'

        img.save(filename)
        print(f"QR code for '{url}' saved as '{filename}'")
        display(Image(filename=filename)) # Display the image in the notebook output
    except Exception as e:
        print(f"Error generating QR code for '{url}': {e}")

print("QR code generator ready.")

while True:
    user_url = input("Enter a URL to generate a QR code (or type 'exit' to quit): ")
    if user_url.lower() == 'exit':
        print("Exiting QR code generator.")
        break
    elif user_url:
        generate_qr_code(user_url)
    else:
        print("No URL entered. Please try again.")
```
